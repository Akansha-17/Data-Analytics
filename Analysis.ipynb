```python
print("\nSummary Statistics:")
print(data.describe())
```

```python
print("\nMean:\n", data.mean())
print("\nMedian:\n", data.median())
print("\nStandard Deviation:\n", data.std())
```

```python
#Check datatypes of the column
print(data.dtypes)
```

```python
#Find mean,mdeian and standard deviation
numeric_data = data.select_dtypes(include=['number'])
print("\nMean:\n", numeric_data.mean())
print("\nMedian:\n", numeric_data.median())
print("\nStandard Deviation:\n", numeric_data.std())
```

```python
#  Interpret 25th and 75th percentiles
print("\n25th Percentile:\n", numeric_data.quantile(0.25))
print("\n75th Percentile:\n", numeric_data.quantile(0.75))
```

```python
import seaborn as sns
```

```python
#Bar Plot
sns.countplot(y=data['Model'])
plt.title('Frequency of Car Models')
plt.xlabel('Count')
plt.ylabel('Car Model')
plt.show()
```

```python

#Histogram
numeric_data = data.select_dtypes(include=['number'])
for col in numeric_data.columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[col], kde=True, bins=20, color='blue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=numeric_data['hp'], y=numeric_data['mpg'], hue=data['Model'])
plt.title('Horsepower vs MPG')
plt.xlabel('Horsepower')
plt.ylabel('Miles Per Gallon (MPG)')
plt.legend(title='Car Name')
plt.show()
```

```python
#Box Plot
# for col in numeric_data.columns:
plt.figure(figsize=(8, 4))
sns.boxplot(data['mpg'])
plt.title(f'Box Plot for mpg')
plt.xlabel('mpg')
plt.show()
```
